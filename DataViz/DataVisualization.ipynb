```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,10,5)
y = x**(1/2)
```

```python
plt.plot(x,y,"g_")
plt.xlabel("x")
plt.ylabel("sqrt(x)")
```

```python
plt.plot(x,y,"r-")
```

```python
plt.subplot(1,2,2)
plt.plot(x,y,"g-")
plt.subplot(1,2,1)
plt.plot(y,x)
```

```python
plt.subplot(2,1,1)
plt.plot(x,y,"g-")
plt.subplot(2,1,2)
plt.plot(y,x)
```

```python
fig = plt.figure()
fig.add_axes([0.1,0.1,0.6,0.8]).plot(y,x,"b+")
```

```python
fig = plt.figure()
fig.add_axes([0.1,0.1,0.9,0.8]).plot(y,x)
fig.add_axes([0.15,0.45,0.5,0.5]).plot(x,y)
```

```python
f , axes = plt.subplots(nrows=1,ncols=2)
axes[0].set_title("X to Power 2")
axes[0].plot(x,y)
axes[1].set_title("X to Power 3")
axes[1].plot(x,x**3)
plt.tight_layout()
```

```python
f = plt.figure(figsize=(4,4))
f.add_axes([0.1,0.1,0.8,0.8]).plot(x,x**2.5)
```

```python
fg , axes = plt.subplots(nrows = 1, ncols=2 , figsize=(8,4))
axes[0].plot(x,x**(1/3))
axes[1].plot(y,x)
axes[0].set_title("G1")
axes[1].set_title("G2")
plt.tight_layout()
```

```python
fg.savefig("plot.png",dpi=150)
```

```python
fig = plt.figure(figsize=(5,5),dpi=100)
axs = fig.add_axes([1,1,1,1])
axs.plot(x,x**2,label="X^2")
axs.plot(x,x**3,label = "X^3")
axs.legend(loc=10)
```
