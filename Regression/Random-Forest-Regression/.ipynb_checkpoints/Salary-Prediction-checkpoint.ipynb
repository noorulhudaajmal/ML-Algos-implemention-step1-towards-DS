# Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Importing dataset

```python
data = pd.read_csv("Position_Salaries.csv")
```

# Data Lookup

```python
sns.pairplot(data)
```

```python
X = data[["Level"]].values
```

```python
y = data[["Salary"]].values
```

# MODEL TRAINING

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
regressor = RandomForestRegressor(n_estimators=10 , random_state=0)
```

```python
regressor.fit(X,y.ravel())
```

# MODEL VISUALISATION

```python
plt.scatter(X,y)
plt.plot(X , regressor.predict(X) , color = "green")
```

```python
plt.scatter(X,y)
X_grid = np.arange(X.min() , X.max() , 0.01)
plt.plot(X_grid , regressor.predict(X_grid.reshape(len(X_grid),1)) , color = "RED")
```

# MODEL EVALUATION
