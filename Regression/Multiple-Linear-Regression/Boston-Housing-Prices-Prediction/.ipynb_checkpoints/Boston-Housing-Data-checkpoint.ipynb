# IMPORTING LIBRARIES

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## READING DATA

```python
data = pd.read_csv("housing.csv",header=None , delimiter = "\s+")
```

## DATA LOOK-UP:

```python
data.head()
```

```python
data.info()
```

#### inserting the columns names instead of integer-heads

```python
cols = ["CRIM","ZN","INDUS","CHAS","NOX","RM","AGE","DIS","RAD","TAX","PTRATIO","B","LSTAT","MEDV"]
```

```python
# CRIM    : CRIME RATE PER CAPITA BY AREA
# ZN      : proportion of residential land zoned for lots over 25,000 sq.ft.
# INDUS   : - proportion of non-retail business acres per town.
# CHAS    : Charles River dummy variable (1 if tract bounds river; 0 otherwise)
# NOX     : nitric oxides concentration (parts per 10 million)
# RM      : average number of rooms per dwelling
# AGE     : proportion of owner-occupied units built prior to 1940
# DIS     : weighted distances to five Boston employment centres
# RAD     : index of accessibility to radial highways
# TAX     : full-value property-tax rate per $10,000
# PTRATIO : pupil-teacher ratio by town
# B       : 1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town
# LSTAT   : PERCENTAGE lower status of the population
# MEDV    : Median value of owner-occupied homes in $1000's
```

##### previous integer-head (column names) 

```python
data.columns
```

##### inserting new column names as describes above in cols list

```python
data.columns = cols
```

## EXPLORATORY DATA ANALYSIS

### CHECKING FOR THE OUTLIERS

```python
figure , axis = plt.subplots(figsize=(16,10),ncols=7,nrows=2)
i=0
axis = axis.flatten()
for col in data.columns:
    sns.boxplot(data=data[col] , ax = axis[i] )
    axis[i].set_title(col)
    i+=1
plt.tight_layout()
```

```python
figure , axis = plt.subplots(figsize=(16,10),ncols=7,nrows=2)
i=0
axis = axis.flatten()
for col in cols:
    sns.histplot(data[col] , kde=True , ax = axis[i] )
    axis[i].set_title(col)
    i+=1
plt.tight_layout()
```

## REFINING THE DATA BY REMOVING THE OUTLIER OF COL = MEDV , THE ONE TO BE PREDICTED!

```python
df = data[data["MEDV"]<50]
```

```python
df.info()
```

```python
figure , axis = plt.subplots(figsize=(16,10),ncols=7,nrows=2)
i=0
axis = axis.flatten()
for col in df.columns:
    sns.histplot(df[col] , kde=True , ax = axis[i] )
    axis[i].set_title(col)
    i+=1
plt.tight_layout()
```

### CHECKING THE CORRELATION BETWEEN DATASET

```python
figure = plt.figure(figsize=(12,8))
sns.heatmap(df.corr(),annot=True)
```

```python
# RAD  |  TAX    >>> 0.91
# MEDV |  LSTAT  >>> 0.76
# MEDV |  PTRATIO  >>> 0.52
# MEDV |  TAX  >>> 0.57
# MEDV |  RM  >>> 0.69
# MEDV |  NOX  >>> 0.52
# MEDV |  INDUS  >>> 0.6
```

#### FROM THIS WE CAN ANALYZE THAT COLUMNS : LSTAT, PTRATIO, TAX,RM,NOX,INDUS CAN BE A GOOD OPTION TO USE TO PREDICT THE MEDV BECAUSE OF HAVING CORRELATION WITH IT >50%

```python
resultant_data = df[["LSTAT","PTRATIO","TAX","RM","NOX","INDUS"]]
y = df["MEDV"]
```

```python
resultant_data.head()
```

## EDA OF REFINED DATASET

```python
figure , axis = plt.subplots(figsize=(12,10),ncols=3,nrows=4)
i=0
axis = axis.flatten()
for col in resultant_data.columns:
    sns.regplot( x = resultant_data[col] , y = y ,  ax = axis[i] )
    axis[i].set_title(col)
    i+=1
for col in resultant_data.columns:
    sns.histplot(resultant_data[col] , kde=True , ax = axis[i] )
    axis[i].set_title(col)
    i+=1
plt.tight_layout()
```

```python
X = resultant_data
```

```python
X.head()
```

```python
figure , axis = plt.subplots(figsize=(12,6),ncols=3,nrows=2)
i=0
axis = axis.flatten()
for col in X.columns:
    sns.histplot(X[col] , kde=True , ax = axis[i] )
    axis[i].set_title(col)
    i+=1
plt.tight_layout()
```

# BUILDING MODEL

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.8,random_state=100)
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
lm.coef_
```

```python
coeff = pd.DataFrame(data=lm.coef_, index = X.columns , columns = ["Coeff"])
```

```python
coeff
```

# PREDICTIONS:

```python
predictions = lm.predict(X_test)
```

```python
plt.scatter(y_test , predictions)
plt.xlabel("y_test")
plt.ylabel("predictions")
```

```python
sns.displot((y_test-predictions),kde=True)
plt.tight_layout
```

# MODEL EVALUATION

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
print(mean_absolute_error(y_test,predictions))
```

```python
print(mean_squared_error(y_test,predictions))
```

```python
print(np.sqrt(mean_squared_error(y_test,predictions)))
```
